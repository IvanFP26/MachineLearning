```python
# Mengimport library yang dibutuhkan
import pandas as pd

# Membaca dataset dan mengubahnya menjadi dataframe
df=pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
# Melihat isi dataframe
df
```

```python
# Melihat lima data teratas
df.head()
```

```python
# Menampilkan informasi detail tentang dataframe seperti jumlah baris, kolom, tipe data dll
df.info()
```

```python
# Menghapus kolom 'User ID' dari DataFrame df
df = df.drop(columns=['User ID'])
df
```

```python
# Melakukan one-hot encoding pada variabel kategori dalam DataFrame
df = pd.get_dummies(df)
df
```

```python
# Mendefinisikan fitur yang akan digunakan untuk modeling
predictions = ['Age' , 'EstimatedSalary' , 'Gender_Female' , 'Gender_Male']
# Membuat sub-DataFrame X berisi hanya fitur yang ditentukan di atas dan y sebagai target prediksi ('Purchased')
X = df[predictions]
y = df['Purchased']
```

```python
# Mengimport library yang dibutuhkan
from sklearn.preprocessing import StandardScaler

# Membuat objek scaler
scaler = StandardScaler()
# Fitting scaler pada data X
scaler.fit(X)
# Menggunakan scaler tersebut untuk mentransformasi data X menjadi skala standar (rata-rata=0, deviasi standar=1)
scaled_data = scaler.transform(X)
# Mengubah array numpy kembali menjadi DataFrame pandas
scaled_data = pd.DataFrame(scaled_data, columns = X.columns)
# Menampilkan lima baris pertama dari data terstandarisasi
scaled_data.head()
```

```python
# Mengimport library yang dibutuhkan
from sklearn.model_selection import train_test_split

# Membagikan data menjadi train dan test set
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
# Mengimport library yang dibutuhkan
from sklearn import linear_model

# Melatih model dengan fungsi fit
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Menguji akurasi model
model.score(X_test, y_test)
```
