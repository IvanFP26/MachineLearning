## Recap
You've built your first model, and now it's time to optimize the size of the tree to make better predictions. Run this cell to set up your coding environment where the previous step left off.

```python
#Nama : Ivan Fernanda Prayoga
#NIM : 1103204035
#Tugas ke 4
#Kaggle Exercise Underfitting and Overvitting

#Memanggil library pandas sebagai pd
import pandas as pd
#memanggil fungsi mean_absolute_error dari library sklearn.metrics
from sklearn.metrics import mean_absolute_error
#memanggil train_test_split dari library sklearn.model_selection
from sklearn.model_selection import train_test_split
#memanggil DecisionTreeRegressor dari library sklearn.tree
from sklearn.tree import DecisionTreeRegressor

#menentukan lokasi file data dan membaca file tersebut menggunakan fungsi read_csv() dari pandas
iowa_file_path = '../input/home-data-for-ml-course/train.csv'
#hasil dari iowa_file_path disimpan di variabel home_data
home_data = pd.read_csv(iowa_file_path)

#mengambil kolom 'SalePrice' dari DataFrame 'home_data' dan menyimpannya dalam variabel 'y'
y = home_data.SalePrice

#mendefinisikan daftar fitur yang akan digunakan dalam model
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
#mengambil kolom-kolom yang ditentukan oleh daftar fitur dari DataFrame home_data dan menyimpannya dalam variabel X
X = home_data[features]

#membagi data menjadi dua set - satu untuk pelatihan dan satu untuk validasi
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

#membuat model regresi tree decission dengan acak ditetapkan ke 1
iowa_model = DecisionTreeRegressor(random_state=1)
#memastikan model menggunakan data
iowa_model.fit(train_X, train_y)

#setelah model ditentukan lalu membuat prediksi menggunakan data validasi
val_predictions = iowa_model.predict(val_X)
#menghitung Mean Absolute Error (MAE)
val_mae = mean_absolute_error(val_predictions, val_y)
#menampilkan hasil output MAE hasil validasi
print("Validation MAE: {:,.0f}".format(val_mae))

#memanggil Dlearntools.core dari library binder
from learntools.core import binder
#memanggil fungsi bind() pada objek binder
binder.bind(globals())
#memanggil semua fungsi dan kelas dari modul ex5 (yaitu, Exercise 5) dalam paket learntools.machine_learning
from learntools.machine_learning.ex5 import *
#menampilkan hasil output setup complete
print("\nSetup complete")
```

# Exercises
You could write the function `get_mae` yourself. For now, we'll supply it. This is the same function you read about in the previous lesson. Just run the cell below.

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```
