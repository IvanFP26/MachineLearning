```python
#Nama : Ivan Fernanda Prayoga
#NIM : 1103204035
#Tugas ke 5
#Kaggle Exercise Random Forest

#Memanggil library pandas sebagai pd
import pandas as pd
#memanggil fungsi mean_absolute_error dari library sklearn.metrics
from sklearn.metrics import mean_absolute_error
#memanggil train_test_split dari library sklearn.model_selection
from sklearn.model_selection import train_test_split
#memanggil DecisionTreeRegressor dari library sklearn.tree
from sklearn.tree import DecisionTreeRegressor


#menentukan lokasi file data dan membaca file tersebut menggunakan fungsi read_csv() dari pandas
iowa_file_path = '../input/home-data-for-ml-course/train.csv'
#hasil dari iowa_file_path disimpan di variabel home_data
home_data = pd.read_csv(iowa_file_path)
#mengambil kolom 'SalePrice' dari DataFrame 'home_data' dan menyimpannya dalam variabel 'y'
y = home_data.SalePrice
#mendefinisikan daftar fitur yang akan digunakan dalam model
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
#mengambil kolom-kolom yang ditentukan oleh daftar fitur dari DataFrame home_data dan menyimpannya dalam variabel X
X = home_data[features]

# Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# Specify Model
iowa_model = DecisionTreeRegressor(random_state=1)
# Fit Model
iowa_model.fit(train_X, train_y)

# Make validation predictions and calculate mean absolute error
val_predictions = iowa_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE when not specifying max_leaf_nodes: {:,.0f}".format(val_mae))

# Using best value for max_leaf_nodes
iowa_model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=1)
iowa_model.fit(train_X, train_y)
val_predictions = iowa_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE for best value of max_leaf_nodes: {:,.0f}".format(val_mae))


#memanggil Dlearntools.core dari library binder
from learntools.core import binder
#memanggil fungsi bind() pada objek binder
binder.bind(globals())
#memanggil semua fungsi dan kelas dari modul ex6 (yaitu, Exercise 6) dalam paket learntools.machine_learning
from learntools.machine_learning.ex6 import *
#menampilkan hasil output setup complete
print("\nSetup complete")
```
