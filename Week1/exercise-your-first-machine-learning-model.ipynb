## Step 2: Create X
Now you will create a DataFrame called `X` holding the predictive features.

Since you want only some columns from the original data, you'll first create a list with the names of the columns you want in `X`.

You'll use just the following columns in the list (you can copy and paste the whole list to save some typing, though you'll still need to add quotes):
  * LotArea
  * YearBuilt
  * 1stFlrSF
  * 2ndFlrSF
  * FullBath
  * BedroomAbvGr
  * TotRmsAbvGrd

After you've created that list of features, use it to create the DataFrame that you'll use to fit the model.

```python
#mendefinisikan dari list feature_names, yang berisi nama-nama fitur yang akan digunakan
feature_names = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF','FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd'] 
#mengambil data dari DataFrame home_data yang sesuai dengan fitur-fitur yang ada dalam feature_names dan menyimpannya dalam variabel X
X = home_data[feature_names] 

#mengecek apakah tulisan sudah benar
step_2.check() 
```

## Review Data
Before building a model, take a quick look at **X** to verify it looks sensible

```python
#menampilkan output dari variabel x
print(X) 
#menampilkan output dari variabel y
print(y) 
```

## Step 3: Specify and Fit Model
Create a `DecisionTreeRegressor` and save it iowa_model. Ensure you've done the relevant import from sklearn to run this command.

Then fit the model you just created using the data in `X` and `y` that you saved above.

```python
#mengimpor kelas DecisionTreeRegressor dari library sklearn
from sklearn.tree import DecisionTreeRegressor 
#membuat sebuah objek model regresi berbasis pohon iowa_model dengan mengatur random_state
iowa_model = DecisionTreeRegressor(random_state=1) 
#menyesuaikan model
iowa_model.fit(X,y) 
#mengecek apakah tulisan sudah benar
step_3.check() 
```

```python
# step_3.hint()
# step_3.solution()
```

## Step 4: Make Predictions
Make predictions with the model's `predict` command using `X` as the data. Save the results to a variable called `predictions`.

```python
#menggunakan model yang telah disesuaikan untuk membuat prediksi berdasarkan X dan menyimpan hasil prediksi dalam variabel predictions
predictions = iowa_model.predict(X) 
#menampilkan output dari variabel predictions
print(predictions) 
#mengecek apakah tulisan sudah benar
step_4.check() 
```
