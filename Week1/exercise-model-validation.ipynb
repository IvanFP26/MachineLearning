```python
#Nama : Ivan Fernanda Prayoga
#NIM : 1103204035
#Tugas ke 3
#Kaggle Exercise Model Validation

#memanggil library pandas sebagai pd
import pandas as pd 
#mengimpor kelas DecisionTreeRegressor dari library sklearn
from sklearn.tree import DecisionTreeRegressor 

#mendefinisikan variabel iowa_file_path yang berisi path dari file CSV
iowa_file_path = '../input/home-data-for-ml-course/train.csv' 
#membaca file CSV yang diberikan oleh iowa_file_path menggunakan pandas
home_data = pd.read_csv(iowa_file_path)
#mengambil kolom SalePrice dari DataFrame home_data dan menyimpannya dalam variabel y
y = home_data.SalePrice 

#mendefinisikan dari list feature_columns, yang berisi nama-nama fitur yang akan digunakan
feature_columns = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
#mengambil data dari DataFrame home_data yang sesuai dengan fitur-fitur yang ada dalam feature_columns dan menyimpannya dalam variabel X.
X = home_data[feature_columns] 

#membuat sebuah objek model regresi berbasis pohon iowa_model
iowa_model = DecisionTreeRegressor()
#menyesuaikan model
iowa_model.fit(X, y) 

#menampilkan nilai target aktual untuk rumah-rumah yang sesuai dengan prediksi pertama
print("First in-sample predictions:", iowa_model.predict(X.head())) 
#menampilkan nilai target yang sebenarnya untuk beberapa data pertama
print("Actual target values for those homes:", y.head().tolist()) 

#mengimpor modul binder dari paket learntools.core
from learntools.core import binder 
#memanggil fungsi yang digunakan untuk menghubungkan sesuatu
binder.bind(globals()) 
#mengimpor semua simbol (fungsi, variabel, dll.) dari modul ex4
from learntools.machine_learning.ex4 import * 
#menampilkan hasil output setup complete
print("Setup Complete") 
```

## Step 1: Split Your Data
Use the `train_test_split` function to split up your data.

Give it the argument `random_state=1` so the `check` functions know what to expect when verifying your code.

Recall, your features are loaded in the DataFrame **X** and your target is loaded in **y**.


```python
#mengimpor fungsi train_test_split dari library sklearn
from sklearn.model_selection import train_test_split 
#membagi data menjadi data (train_X dan train_y) dan data validasi (val_X dan val_y) dengan menggunakan train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
#mengecek apakah tulisan sudah benar
step_1.check() 
```

```python
# The lines below will show you a hint or the solution.
# step_1.hint() 
# step_1.solution()
```

## Step 2: Specify and Fit the Model

Create a `DecisionTreeRegressor` model and fit it to the relevant data.
Set `random_state` to 1 again when creating the model.

```python
#membuat sebuah objek model regresi berbasis tree iowa_model dengan mengatur random_state
iowa_model = DecisionTreeRegressor(random_state=1) 
#memastikan model iowa_model dengan data fit
iowa_model.fit(train_X, train_y) 
#mengecek apakah tulisan sudah benar
step_2.check() 
```

```python
# step_2.hint()
# step_2.solution()
```

## Step 3: Make Predictions with Validation data


```python
#membuat prediksi pada data validasi val_X dan menyimpan hasil prediksi dalam val_predictions
val_predictions = iowa_model.predict(val_X) 
#mengecek apakah tulisan sudah benar
step_3.check() 
```

```python
# step_3.hint()
# step_3.solution()
```

Inspect your predictions and actual values from validation data.

```python
#mencetak lima nilai pertama dari hasil prediksi (val_predictions) pada data 
print(val_predictions[:5])
#mencetak lima nilai pertama dari target prediksi yang sebenarnya (val_y) pada data 
print(val_y[:5])
```

What do you notice that is different from what you saw with in-sample predictions (which are printed after the top code cell in this page).

Do you remember why validation predictions differ from in-sample (or training) predictions? This is an important idea from the last lesson.

## Step 4: Calculate the Mean Absolute Error in Validation Data


```python
#mengimpor fungsi mean_absolute_error dari library sklearn
from sklearn.metrics import mean_absolute_error
#menghitung MAE antara nilai target sebenarnya (val_y) dan hasil prediksi (val_predictions) pada data 
val_mae = mean_absolute_error(val_y, val_predictions)
#menampilkan output val_mae
print(val_mae)
#mengecek apakah tulisan sudah benar
step_4.check()
```
