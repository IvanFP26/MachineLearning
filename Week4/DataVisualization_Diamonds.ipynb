```python
# Mengimpor library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from ipywidgets import interact, widgets
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Menggunakan path ke file di Google Drive
file_path = '/content/drive/MyDrive/Machine Learning/diamonds.csv'

# Membaca file ke dalam DataFrame
df = pd.read_csv(file_path)
print(df.head())
```

```python
fitur = ['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'price', 'x', 'y', 'z']

# Loop melalui setiap fitur dan mencetak jumlah nilai unik
for col in fitur:
    unique_count = df[col].nunique()  # Menghitung jumlah nilai unik dalam fitur
    print(f"Jumlah unik untuk '{col}': {unique_count}")
```

```python
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Daftar fitur kategoris (categorical features)
categorical_features = ['cut', 'color', 'clarity']

# Fitur target
target_feature = 'price'

# Memisahkan fitur menjadi numerical, categorical, dan target
numerical_df = df[numerical_features]
categorical_df = df[categorical_features]
target_df = df[target_feature]
print(numerical_df)
print(categorical_df)
print(target_df)
```

```python
categorical_features = ['cut', 'color', 'clarity']

# Loop melalui setiap fitur kategoris dan membuat bar chart
for feature in categorical_features:
    # Menghitung jumlah masing-masing nilai dalam fitur kategoris
    value_counts = df[feature].value_counts()

    # Membuat bar chart
    plt.figure(figsize=(8, 6))
    value_counts.plot(kind='bar', color='skyblue')
    plt.title(f'Bar Chart for {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)  # Untuk memutar label sumbu x agar lebih mudah dibaca
    plt.show()
```

```python
categorical_features = ['cut', 'color', 'clarity']

# Definisi palet warna gelap
dark_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']

# Loop melalui setiap fitur kategoris dan membuat pie chart dengan palet warna gelap
for index, feature in enumerate(categorical_features):
    # Menghitung jumlah masing-masing nilai dalam fitur kategoris
    value_counts = df[feature].value_counts()

    # Membuat pie chart dengan menggunakan palet warna gelap
    plt.figure(figsize=(8, 6))
    plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140, colors=dark_colors)
    plt.title(f'Pie Chart for {feature}')
    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.show()
```

```python
# Mengelompokkan numerical features
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Membuat distribusi untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature], kde=True, color='darkcyan')
    plt.title(f'Distribution of {feature}', size=15)
    plt.xlabel(feature, size=12)
    plt.ylabel('Frequency', size=12)
    plt.show()
```

```python
n_rows = 2
n_cols = 3

# Membuat subplots untuk menampilkan distribusi
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))

for i, feature in enumerate(numerical_features):
    row = i // n_cols
    col = i % n_cols

    # Histogram plot
    sns.histplot(df[feature], kde=True, ax=axes[row][col])
    axes[row][col].set_title(f'Histogram for {feature}')

# Memastikan tidak ada subplot yang kosong
for i in range(len(numerical_features), n_rows * n_cols):
    row = i // n_cols
    col = i % n_cols
    fig.delaxes(axes[row][col])

plt.tight_layout()
plt.show()
```

```python
for feature in numerical_features:
    fig = px.histogram(df, x=feature, title=f'Histogram for {feature}')

    # Menambahkan kontrol bin width yang dapat diubah
    fig.update_traces(marker_line_width=0.2, marker_line_color="black")
    fig.update_layout(bargap=0.1)
    fig.update_xaxes(categoryorder="total ascending")

    fig.show()
```

```python
n_rows = 2
n_cols = 3

# Membuat subplot untuk menampilkan boxplot
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10), sharex=True)

for i, feature in enumerate(numerical_features):
    row = i // n_cols
    col = i % n_cols

    # Menggambar boxplot dengan orientasi horizontal
    sns.boxplot(x=df[feature], ax=axes[row][col], orient='h', color='skyblue')
    axes[row][col].set_title(f'Box Plot for {feature}')

# Memastikan tidak ada subplot yang kosong
for i in range(len(numerical_features), n_rows * n_cols):
    row = i // n_cols
    col = i % n_cols
    fig.delaxes(axes[row][col])

# Menampilkan grafik
plt.tight_layout()
plt.show()
```
